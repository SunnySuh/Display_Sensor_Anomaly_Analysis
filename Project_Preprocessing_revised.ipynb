## 다중공선성 문제 - 팀원끼리 컬럼 나눠서 실행해봄

```python
# 상관계수가 1인 컬럼 목록 파일 불러오기
#corr_df = pd.read_csv('/content/drive/MyDrive/나무플래닛/3. 산출물/same_coef_frame.xlsx',encoding='utf-8')

# 담당한 컬럼명 리스트 만들기
corr_col = ['ATM.FLOW.SECT.6.N2.H2.1FI33410.PV',
'ATM.FLOW.SECT.7.N2.H2.1FI33412.PV',
'ATM.FLOW.SECT.9.N2.H2.1FIC33416.PV',
'ATM.FLOW.SECT.10.N2.H2.1FIC33418.PV']
corr_col = list(set(corr_col))
x ='ATM.FLOW.SECT.6.N2.H2.1FI33410.PV'
corr_col.remove(x)
corr_col
```

```python
# 전체 컬럼 중 해당 집단의 컬럼명만 제외
# 전체 컬럼명 리스트
columns_all = data.columns.to_list()
# 확인
print(len(columns_all))

# date와 폐기율 컬럼 제외 #'투입.LEFT.','투입.RIGHT.'??
drop_columns = ['투입.LEFT.','투입.RIGHT.','ROLLDOWN.폐기.LEFT.','진공패드.폐기.LEFT.','ROLLDOWN.폐기.RIGHT.','진공패드.폐기.RIGHT.','폐기율...','ROLLDOWN.진공패드.폐기율..LEFT.','ROLLDOWN.진공패드.폐기율..RIGHT.','ROLLDOWN.진공패드.폐기율..전체.','L.RD','L.Vac','R.RD','R.Vac']

for x in drop_columns:
  columns_all.remove(x)

# 확인
print(len(columns_all))

# 전체 컬럼 중 해당 집단의 컬럼명만 제외
for x in corr_col:
  columns_all.remove(x)
# 확인
print(len(columns_all))
```

```python
,,,,,,,,,,,# 컬럼 간 다중공선성을 확인하자
# 분산팽창요인(VIF) > 10 이면 다중공선성 문제가 발생한 것임
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 피처마다의 VIF 계수를 출력
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(data[columns_all].values, i) for i in range(data[columns_all].values.shape[1])]
vif["features"] = data[columns_all].columns
vif
```

```python
## VIF 파일 저장
vifdf = vif.sort_values(by="VIF Factor",ascending=False).round(1)
vifdf.to_csv('/content/drive/MyDrive/Colab Notebooks/Display_Sensor_Anomaly_Analysis/vifdf_yh.csv')

vifdf = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Display_Sensor_Anomaly_Analysis/vifdf_yh.csv', usecols=[1,2])
vifdf
```
